Using the methods provided in archiDART, simulations produced by various root architecture models can be compared in three main ways:

1. **_Qualitatively_**, using advanced plotting functions provided in the ggplot2 R package;

2. **_Quantitatively_**, by computing traits at the plant or root level;

3. **_Quantitatively_**, by analysing the topology of plant root systems using persistent homology.

In the next sections, we will describe how root system simulations can be compared using the approaches listed above.

We will also consider the following hypotheses:

- A total of **m** root architecture models have to be compared;
- Each root architecture model is used to model **x** different scenarios;
- Each scenario is simulated **n** times with each root architecture model;
- For each root architecture model, all simulations produced for a given scenario are stored into one directory (Scenario --> Model --> RSMLs).

## 3. First approach: using plotting functions to qualitatively compare root systems

archiDART contains several graphical tools to plot root systems, but when data are stored in a rsmlToTable object, the functions of the ggplot2 package are certainly the best option to display root systems and visually compare outputs from different root architecture models.

Root segments can be plotted using a colour code based on one of the following variables:
- **time**: the emergence time
- **order**: the branching order
- **x2/y2/z2**: the depth (choose the axis having the same direction as the gravity vector)
- **diameter2**: the root diameter (only if diameter available)
- **surface**: the root surface area (only if diameter available)
- **volume**: the root volume (only if diameter available)
- **orientation**: the orientation relative to a vertical direction vector
- **growth**: the growth rate
- **geodesic**: the geodesic distance (i.e., the distance measured along the roots between the root system base and any point of the root system)

To change the colour code used for plotting, just change the `colour`argument of the `aes` function in the following code.

```python
#If the models do not have the same vertical axis
ggplot(table21)+
geom_segment(aes(x=x1, y=-y1, xend=x2, yend=-y2, colour=geodesic), size=0.5, alpha=1)+
coord_fixed()+
theme_bw()+
facet_wrap(~file, ncol=5)+
xlab(paste("X (", unit, ")", sep=""))+
ylab(paste("Y (", unit, ")", sep=""))

ggplot(table22)+
geom_segment(aes(x=x1, y=z1, xend=x2, yend=z2, colour=geodesic), size=0.5, alpha=1)+
coord_fixed()+
theme_bw()+
facet_wrap(~file, ncol=5)+
xlab(paste("X (", unit, ")", sep=""))+
ylab(paste("Z (", unit, ")", sep=""))

#If the models have the same vertical axis
#ggplot(table2)+
#geom_segment(aes(x=x1, y=-y1, xend=x2, yend=-y2, colour=geodesic), size=0.5, alpha=1)+
#coord_fixed()+
#theme_bw()+
#facet_wrap(~file, ncol=5)+
#xlab(paste("X (", unit, ")", sep=""))+
#ylab(paste("Y (", unit, ")", sep=""))
```

## 4. Second approach: computing traits at the plant or root level to compare root systems

### 4.1. At the plant level

In archiDART, the `architect` function can be used to compute traits at the plant level (i.e., for the entire root system). For each root system and each observation date, the following parameters will be calculated:

- **TRL**: total root length
- **GRTR**: growth rate of the root system
- **L1R**: first-order root length
- **GR1R**: first-order root growth rate
- **TN1R**: total number of first-order roots
- **TNLR**: total number of lateral roots
- **TLRL**: total lateral root length
- **NxLR**: total number of lateral roots with a branching order equal to x
- **LxLR**: total length of lateral roots with a branching order equal to x
- **MLxLR**: mean length of lateral roots with a branching order equal to x
- **GRxL**: growth rate of roots with a branching order equal to x
- **D2LR**: density of secondary roots on first-order roots
- **Height**: root system height
- **Width**: root system width
- **Convexhull**: area of the convex hull

If diameter values are provided in the RSML files, the following parameters will be calculated for each root system at the last observation date:

- **MDx**: mean diameter of roots with a branching order equal to x
- **MDLR**: mean lateral root diameter
- **Sx**: total surface area of roots with a branching order equal to x
- **Stot**: total root surface area
- **Vx**: total volume of roots with a branching order equal to x
- **Vtot**: the root system volume

The rsmlToTable objects created earlier can directly be used as an input of the `architect` function. Root traits can be calculated using the following code:

```python
#If the models do not have the same vertical axis
plant11<-architect(inputrsml=table11, vertical3d=vertical1)
plant12<-architect(inputrsml=table12, vertical3d=vertical2)
plant21<-architect(inputrsml=table21, vertical3d=vertical1)
plant22<-architect(inputrsml=table22, vertical3d=vertical2)

#Create 1 data frame for each scenario
plant1<-smartbind(plant11, plant12)
plant2<-smartbind(plant21, plant22)

#If the models have the same vertical axis
#plant1<-architect(inputrsml=table1)
#plant2<-architect(inputrsml=table2)
```

```python
#Add 1 column for model and 1 column for replicate
plant1$model <- unlist(lapply(strsplit(as.character(plant1$FileName), "-"), `[[`, 1))[]
plant2$model <- unlist(lapply(strsplit(as.character(plant2$FileName), "-"), `[[`, 1))[]
plant1$replicate <- unlist(lapply(strsplit(as.character(plant1$FileName), "-"), `[[`, 2))[]
plant2$replicate <- unlist(lapply(strsplit(as.character(plant2$FileName), "-"), `[[`, 2))[]
plant1$replicate <- unlist(lapply(strsplit(as.character(plant1$replicate), "_"), `[[`, 1))[]
plant2$replicate <- unlist(lapply(strsplit(as.character(plant2$replicate), "_"), `[[`, 1))[]
head(plant1)
```

**For each scenario, we can now compare outputs of different root architecture models.** Two different strategies will be described:

- **Univariate analysis**: simulations of different root architecture models are compared one variable at a time;
- **Multivariate analysis**: selected variables calculated by `architect` are used to compare root architecture models.

**First, let's have a look at the univariate approach:**

You can choose the variable you want to plot by changing the `y` argument of the `aes` function.

```python
#Plot results for all simulations (Scenario 1)
ggplot(plant2)+
geom_line(aes(x=Time, y=TRL, colour=model, group=FileName), size=1)+
theme_bw()
```

```python
#Plot average per simulation (Scenario 1)
ggplot(plant2)+
stat_smooth(aes(x=Time, y=TRL, colour=model), size=1)+
theme_bw()
```

```python
#Beanplots are a convenient way to compare outputs of different root architecture models for specific observation dates
#Choose an observation date
time<-20
sub<-plant2[plant2$Time==time,]

#Choose the colors used for plotting (area of the beans, lines inside the bean, lines outside the bean, average line per bean)
colors<-c("white", "black", "black", "red")

#Beanplot
par(mfrow=c(2,2))
gf1<-beanplot(TRL~model, sub, log="", bw="nrd0", border = NA, las=1, col=colors, ylab="Total root length", what=c(1,0,1,1))
gf2<-beanplot(L1R~model, sub, log="", bw="nrd0", border = NA, las=1, col=colors, ylab="First-order root length", what=c(1,0,1,1))
gf3<-beanplot(TNLR~model, sub, log="", bw="nrd0", border = NA, las=1, col=colors, ylab="Total lateral root number", what=c(1,0,1,1))
gf4<-beanplot(Convexhull~model, sub, log="", bw="nrd0", border = NA, las=1, col=colors, ylab="Convexhull surface area", what=c(1,0,1,1))
```

**Simulations produced by root architecture models can also be compared using multivariate statistical tools (such as principal component analysis or discriminant analysis).**

```python
#Choose an observation date
time<-20
sub<-plant2[plant2$Time==time,]
rownames(sub)<-sub$FileName

#Select variables for PCA
datapca<-sub[,c("TRL", "L1R", "TN1R", "TNLR", "TLRL", "N2LR", "L2LR", "D2LR", "Height", "Width", "Convexhull", "model")]

#PCA (! quali.sup)
pca<-PCA(datapca, scale.unit=TRUE, ncp=3, quali.sup=12)
indpca<-as.data.frame(pca$ind$coord)
indpca<-as.data.frame(lapply(indpca, function(x) as.numeric(as.character(x))))
indpca$model<-sub$model
pca$var$cor
```

```python
#Plot PCA
ggplot(indpca)+
geom_point(aes(Dim.1, Dim.2, colour=model))+
xlab(paste("PC1 (", round(pca$eig[1,2], 1), "%)", sep=""))+
ylab(paste("PC2 (", round(pca$eig[2,2], 1), "%)", sep=""))+
geom_hline(yintercept=0)+
geom_vline(xintercept=0)+
theme_bw()
```

### 4.2. At the root level

In archiDART, the `root` function can be used to compute traits/attributes at the root level (i.e., for each individual root). For each root, the following parameters will be calculated:

- **order**: root branching order
- **DBase**: distance between the branching point and the parent root base
- **length**: root length
- **nlat**: number of lateral roots
- **branching.angle**: the branching (or insertion) angle
- **tortuosity**: the ratio of the final root length to the Euclidean distance between the branching point and the apex of the root
- **growth**: root growth rate
- **lauz**: length of unbranched apical zone (see [Pags et al (2010)](https://link.springer.com/article/10.1007/s11104-009-0079-x))

If diameter values are provided in the RSML files, the following parameters will also be calculated for each root:

- **mean.diameter**: mean diameter of the root
- **sd.diameter**: standard deviation of the root diameter
- **surface**: total root surface area
- **volume**: total root volume

The rsmlToTable objects created earlier can directly be used as an input of the `root` function. Root traits/attributes can be calculated using the following code:

```python
#If the models do not have the same vertical axis
#last=TRUE if only interested in comparing roots at the last observation date
#last=FALSE computes all parameters at each observation date
root11<-root(x=table11, vertical3d=vertical1, last=TRUE, show.progress=TRUE)
root12<-root(x=table12, vertical3d=vertical2, last=TRUE, show.progress=TRUE)
root21<-root(x=table21, vertical3d=vertical1, last=TRUE, show.progress=TRUE)
root22<-root(x=table22, vertical3d=vertical2, last=TRUE, show.progress=TRUE)

#Create 1 data frame for each scenario
root1<-smartbind(root11, root12)
root2<-smartbind(root21, root22)

#If the models have the same vertical axis
#root1<-root(x=table1, last=TRUE, show.progress=TRUE)
#root2<-root(x=table2, last=TRUE, show.progress=TRUE)
```

```python
#Add 1 column for model and 1 column for replicate
root1$model <- unlist(lapply(strsplit(as.character(root1$file), "-"), `[[`, 1))[]
root2$model <- unlist(lapply(strsplit(as.character(root2$file), "-"), `[[`, 1))[]
root1$replicate <- unlist(lapply(strsplit(as.character(root1$file), "-"), `[[`, 2))[]
root2$replicate <- unlist(lapply(strsplit(as.character(root2$file), "-"), `[[`, 2))[]
head(root1)
```

**We can now compare root traits/attributes for different root types using the following code:**

```python
#Beanplots are a convenient way to compare outputs of different root architecture models for specific observation dates
#!!! last observation date is different for archisimple and crootbox in this example.
#Makes more sense to compare models with same simulation time.

#Choose the colors used for plotting (area of the beans, lines inside the bean, lines outside the bean, average line per bean)
colors<-c("white", "black", "black", "red")

#Beanplot
par(mfrow=c(2,2))
gf1<-beanplot(length~model*order, root2, log="", bw="nrd0", border = NA, col=colors,
              ylab="Root length", las=2, what=c(1,0,1,1))
gf2<-beanplot(nlat~model*order, root2, log="", bw="nrd0", border = NA, col=colors,
              ylab="Number of laterals", las=2, what=c(1,0,1,1))
gf3<-beanplot(branching.angle~model*order, root2, log="", bw="nrd0", border = NA, col=colors,
              ylab="Branching angle", las=2, what=c(1,0,1,1))
gf4<-beanplot(lauz~model*order, root2, log="", bw="nrd0", border = NA, col=colors,
              ylab="Length apical unbranched zone", las=2, what=c(1,0,1,1))
```

## 5. Third approach: topological analysis of root systems using persistent homology

Persistent homology is a mathematical framework allowing the topological analysis of branching structures, such as shoot and root systems. Recently, a new data analysis pipeline was implemented in archiDART to analyse root system architectures using persistent homology. More information can be found in [Li et al (2017)](https://onlinelibrary.wiley.com/doi/abs/10.3732/ajb.1700046) and [Delory et al (2018)](https://f1000research.com/articles/7-22/v1).

After package installation, the topological analysis of plant root systems (RSML files) using persistent homology comprises four main steps: (1) creating an rsmlToTable object, (2) computing persistence barcodes, (3) computing a pairwise bottleneck distance matrix, and (4) visualizing topological differences between root systems using multidimensional scaling (MDS).

Because archiDART relies on another package (TDA) that was recently archived by CRAN maintainers (i.e., it is not available anymore on CRAN repository), **you need to install it manually before analysing your RSML files**. You can do it easily using the following code in your R console:

```python
#install.packages("drat")
#drat::addRepo("archidart")
#install.packages("TDA")
```

Now that all dependencies are installed, you can start analysing your data!

**First, compute the persistence barcodes using a geodesic distance function.**

The main output of a persistent homology analysis is a persistence barcode recording the birth (apparition of a new connected component) and death (fusion of two connected components) of each branch when a distance function traverses the branching structure. For a given mathematical function, one persistence barcode can be computed for each simulated root system. Look at Figure 1 and interactive Figure 3 [in this paper](https://f1000research.com/articles/7-22/v1) for a more accurate description of a persistence barcode. Barcodes can be computed using the following code:

```python
ph<-perhomology(table2, show.progress=TRUE)
```

**Here is how a persistence barcode looks like:**

```python
plot(ph[[1]], las=1, main=names(ph)[1])
```

**To compare persistence barcodes against each other, a pairwise distance matrix is needed and the bottleneck distance is one possible option.** The bottleneck distance is considered as a robust dissimilarity metric between two persistence barcodes, and its interpretation is quite straightforward: the greater the distance between two persistence barcodes, the greater will be the dissimilarity between them. Such pairwise bottleneck distance matrix can be calculated with the `bottleneckdist` function of the package. This function only requires a perhomology object as an input. It has to be noted that the computation time required to compute a bottleneck distance matrix is highly dependent on the number and complexity of root systems being compared.

```python
library(TDA)
distance<-bottleneckdist(ph, show.progress=TRUE)
```

**Dissimilarities between root systems can then be visualised using multidimensional scaling:**

```python
#Multidimensional scaling
#k is the number of dimensions
k<-3
mds1<-cmdscale(distance, k=k, eig=TRUE, add=TRUE)

#Evaluate quality of representation using a Shepard plot
realdist<-as.vector(as.dist(distance))
confdist<-as.vector(dist(mds1$points))
plot(realdist, confdist, xlab="Real distances", ylab="Distances in configuration", pch=16, main="Shepard plot", las=1)
abline(a=0, b=1, col="red")
```

```python
#Plot MDS
model<-unlist(lapply(strsplit(as.character(rownames(mds1$points)), "-"), `[[`, 1))[]
datamds<-data.frame(mds1$points, model=model)
colnames(datamds)<-c(paste(rep("MDS", k), c(1:k), sep=""), "model")

ggplot(datamds)+
geom_point(aes(MDS1, MDS2, colour=model))+
geom_hline(yintercept=0)+
geom_vline(xintercept=0)+
ggtitle("Metric multidimensional scaling")+
theme_bw()
```

**Topological differences between branching structures can be further investigated using PERMANOVA** (for PERmutational Multivariate ANalysis Of VAriance). If several root architecture models are compared, the following code can be used to test for statistical differences between groups using a bottleneck distance matrix as an input:

```python
#PERMANOVA
permanova<-adonis(as.dist(distance)~model, data=data.frame(file=rownames(distance), model=model), permutations = 10000,
                  sqrt.dist=TRUE)
permanova
```

```python
#Look at permutations
densityplot(permustats(permanova))
```

```python
#Chech PERMANOVA assumption (multivariate homogeneity of group dispersions)
bd<-betadisper(as.dist(distance), model)
bd
anova(bd)
permutest(bd, permutations=100)
```
